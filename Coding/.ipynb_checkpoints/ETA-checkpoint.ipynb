#### obesity

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r'C:\Users\Jashwanth\OneDrive\Documents\MEDPREDICT\Datasets/ObesityDataSet_raw_and_data_sinthetic.csv')  
df.head()
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns

label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col])

transformed_file_path = "C:/Users/Jashwanth/OneDrive/Documents/MEDPREDICT/Cleaned Datasets/transformed_ObesityDataSet.csv"  
df.to_csv(transformed_file_path, index=False)

print("Transformation complete. File saved at:", transformed_file_path)
```
